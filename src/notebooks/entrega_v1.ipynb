```python
# Generacion del dataframe de palabras:
diccionario_mesa = {'palabra':[], 'linea_estela':[], 'numero_ oracion':[]}
#l铆nea 1:
numero_oracion = 1
for i in linea_1:
    diccionario_mesa['palabra'].append(i)
    diccionario_mesa['linea_estela'].append(1)
    diccionario_mesa['numero_ oracion'].append(numero_oracion)

# Resto de estela:
for fila in range(1, len(mesa)-1):
    linea = [i for i in mesa.loc[fila, 'Texto'].split() if i != '']
    for palabra in linea:
        if palabra == '|':
            numero_oracion += 1
        else:
            palabra = palabra.replace('[', '')
            palabra = palabra.replace(']', '')
            palabra = palabra.replace('.', '')
            palabra = palabra.replace('', '')
            diccionario_mesa['palabra'].append(palabra)
            diccionario_mesa['linea_estela'].append(fila+1)
            diccionario_mesa['numero_ oracion'].append(numero_oracion) 


df_palabras = pd.DataFrame(diccionario_mesa)
df_palabras[:27]


```

```python
# Vamos a hacer que se pueda leer:
import math
def feniciohebreo_a_latino(palabra):
    
    # if math.isnan(palabra):
    #     return palabra
    # else:
    try:
        string = []
        for i in palabra:
            if i == '' or i == '':
                string.append('示')
            elif i == '' or i == '':
                string.append('b')
            elif i == '' or i == '':
                string.append('g')
            elif i == '' or i == '':
                string.append('d')
            elif i == '' or i == '':
                string.append('h')
            elif i == '' or i == '':
                string.append('w')
            elif i == '' or i == '':
                string.append('z')
            elif i == '' or i == '':
                string.append('岣')
            elif i == '' or i == '':
                string.append('峁')
            elif i == '' or i == '':
                string.append('y')
            elif i == '' or i == '' or i == '':
                string.append('k')
            elif i == '' or i == '':
                string.append('l')
            elif i == '' or i == '' or i == '':
                string.append('m')
            elif i == '' or i == '' or i == '':
                string.append('n')
            elif i == '' or i == '住':
                string.append('s')
            elif i == '' or i == '注':
                string.append('士')
            elif i == '' or i == '祝' or i == '驻':
                string.append('p')
            elif i == '' or i == '抓' or i == '爪':
                string.append('峁')
            elif i == '' or i == '拽':
                string.append('q')
            elif i == '' or i == '专':
                string.append('r')
            elif i == '' or i == '砖':
                string.append('拧')
            elif i == '' or i == '转':
                string.append('t')
            else:
                print('Caracter no reconocido')
                return 0
        return ''.join(string)
    except:
        print('Caracter no iterable')
        return 0
        

```

```python
import numpy as np
def dibujar_zipf_at(df, columna, colorin, titulo):
    frecuencias = df[columna].sort_values(ascending=False)
    rangos = range(1,len(df[columna].sort_values())+1)
    m, k = np.polyfit(np.log10(rangos), np.log10(frecuencias), 1)
    plt.loglog(rangos, frecuencias, 'o', color =colorin)
    ajuste = (10**k)/(np.array(rangos))**(-1 * m)
    plt.loglog(rangos, ajuste, color ='blue')
    plt.xlabel('Posici贸n del ranking')
    plt.ylabel('Frecuencia')
    plt.title(titulo)
    plt.show()
    # Despu茅s de dibujar, devuelve la pendiente de la recta ajustada por m铆nimos cuadrados
    return m

dibujar_zipf_at(biblia, 'occurrences', 'red', 'Zipf AT')
```
